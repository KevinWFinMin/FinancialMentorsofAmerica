```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('FteEnrollStateDistrict17.csv')
```

```python
df1 = df[(df['COURSE TITLE']=='FINANCIAL MATHEMATICS')|
         (df['SUBJECT']=='FINANCE')|
         (df['COURSE TITLE']=='PERSONAL FINANCIAL LITERACY')]
```

```python
df1['COURSE TITLE'].value_counts()
```

```python
df1 = df1[((df1['COURSE TITLE']=='MONEY MATTERS')|
   (df1['COURSE TITLE']=='FINANCIAL MATHEMATICS')|
   (df1['COURSE TITLE']=='PERSONAL FINANCIAL LITERACY')|
   (df1['COURSE TITLE']=='FINANCIAL ANALYSIS'))&
    (df1['COURSE ENROLLMENT']>0)]
```

```python
df1['COURSE ENROLLMENT'].describe()
```

```python
sns.distplot(df1['COURSE ENROLLMENT'],kde=False)
```
